```python
train_pbs = train_data['problem_id'].unique()
test_pbs = test_data['problem_id'].unique()
val_pbs = val_data['problem_id'].unique()
```

```python
#find intersection between train and test pbs
intersection = set(train_pbs).intersection(test_pbs)
print(len(intersection))
intersection = set(test_pbs).intersection(val_pbs)
print(len(intersection))
print(len(set(train_pbs).intersection(val_pbs)))
```

```python
all_pbs = set(train_pbs).union(test_pbs).union(val_pbs)
```
