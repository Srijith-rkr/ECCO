```python
import pandas as pd
import os
import glob

tag="pretrained"
# Directory containing the JSONL files
directory = f"/data/tir/projects/tir6/general/swaghjal/nl2code_eval_results/percentile_results/{tag}"

# Get a list of all JSONL files in the directory
jsonl_files = glob.glob(os.path.join(directory, "*.jsonl"))

# Initialize lists to store results
file_names = []
num_records = []
runtime_percentile_means = []
mem_percentile_means = []

# Process each file
for file in jsonl_files:
    try:
        # Load the data
        percen_data = pd.read_json(file, orient='index')
        original_length = len(percen_data)
        
        # Drop rows where 'runtime_percentile' is NaN
        percen_data = percen_data.dropna(subset=['runtime_percentile'])
        
        # Collect results
        file_names.append(os.path.basename(file))
        num_records.append(len(percen_data))
        runtime_percentile_means.append(percen_data['runtime_percentile'].mean())
        mem_percentile_means.append(percen_data['mem_percentile'].mean())
        
        # Print results for the current file
        print(f"File: {os.path.basename(file)}")
        print(f"Number of records: {len(percen_data)}")
        print(f"original_length: {original_length}")
        print(f"Pass@1:{(len(percen_data)/original_length)*100}")
        print(f"Mean runtime percentile: {percen_data['runtime_percentile'].mean()*100}")
        print(f"Mean memory percentile: {percen_data['mem_percentile'].mean()*100}")
        print("\n")
    
    except Exception as e:
        print(f"Failed to process file {file} due to {e}")

# Summary DataFrame
summary_df = pd.DataFrame({
    'file_name': file_names,
    'num_records': num_records,
    'runtime_percentile_mean': runtime_percentile_means,
    'mem_percentile_mean': mem_percentile_means
})

# Print the summary DataFrame
# print(summary_df)
```
