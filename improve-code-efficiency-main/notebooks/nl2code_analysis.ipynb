```python
train_pbs = list(train_data['problem_id'].unique())
```

```python
len(train_pbs)
```

```python
#Filter trajectories based on train_pbs
trajectories = trajectories[trajectories['problem_id'].isin(train_pbs)]
```

```python
trajectories
```

```python
problem_descriptions = pd.read_json('/data/tir/projects/tir6/general/swaghjal/nl_prob_desc_pie_dataset.jsonl', lines=True, orient='records')
```

```python
problem_descriptions
```

```python
#Filter problem descriptions based on train_pbs
trajectories_pbs = list(trajectories['problem_id'].unique())
problem_descriptions = problem_descriptions[problem_descriptions['problem_id'].isin(trajectories_pbs)]
```

```python
len(problem_descriptions)
```

```python
def get_fastest_code(row):
    min_index = row['cpu_time'].index(min(row['cpu_time']))
    return row['trajectories'][min_index]

# Apply the function to create the 'fastest_code' column
trajectories['fastest_code'] = trajectories.apply(get_fastest_code, axis=1)
```

```python
trajectories.reset_index(drop=True, inplace=True)
```

```python
trajectories
```

```python
#Create a new dataframe that combines problem id, problem description, and fastest code
combined_data = pd.merge(trajectories, problem_descriptions, on='problem_id')
```

```python
#Take only three columns of combined_data problem id, problem description, and fastest code
combined_data = combined_data[['problem_id', 'problem_description', 'fastest_code']]
```
