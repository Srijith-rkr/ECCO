````python
from tqdm import tqdm

filtered_data = []

# Iterate over each row in the DataFrame
for index, row in tqdm(user_trajectories.iterrows(), total=len(user_trajectories)):
    # Get the indexes where status is "Accepted"
    accepted_indexes = [i for i, status in enumerate(row['status']) if status == 'Accepted']
    if len(accepted_indexes) < 3:
        continue
    
    # Filter the data for the accepted indexes
    filtered_row = {
        'user_id': row['user_id'],
        'problem_id': row['problem_id'],
        'original_language': row['original_language'],
        'trajectories': [row['trajectories'][i] for i in accepted_indexes],
        'submission_id': [row['submission_id'][i] for i in accepted_indexes],
        'status': [row['status'][i] for i in accepted_indexes],
        'cpu_time': [row['cpu_time'][i] for i in accepted_indexes],
        'memory': [row['memory'][i] for i in accepted_indexes],
        'code_size': [row['code_size'][i] for i in accepted_indexes],
        'accuracy': [row['accuracy'][i] for i in accepted_indexes]
    }
    filtered_data.append(filtered_row)

# Convert the filtered_data to a DataFrame
filtered_df = pd.DataFrame(filtered_data)
````

````python
filtered_df.drop('accuracy',axis=1,inplace=True)
````

````python
def get_sorted_indexes_after_filtering(input_list):
    # Filter the list in ascending order
    filtered_list = sorted(input_list, reverse=True)

    # Get the sorted indexes after filtering
    sorted_indexes = [input_list.index(num) for num in filtered_list]

    return sorted_indexes

def update_list_in_sequence(input_list, sorted_indexes):
    # Create a new list to store the updated elements
    updated_list = []

    # Update the list in the same sequence as the sorted indexes
    for idx in sorted_indexes:
        updated_list.append(input_list[idx])

    return updated_list
````
